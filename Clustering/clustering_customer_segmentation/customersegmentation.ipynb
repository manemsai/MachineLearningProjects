

                                Mall Customer Segmentation
##### Objective:
Dividing the customers into five various groups.
Clustering based on 2 features (Annual Income in k$  and Spending Score) in order to visulaize or plot the clusters


##### Dataset Description:
        

This Dataset has complete the information about Mall Customer Spending Score.
* The Dataset contains five attributes and 200 instances.
* The 1st attribute is Customer ID which has every Customer has Unique
* The 2nd attribute is Gender which is of course male/female.
* The 3rd attribute is age which is between 19 to 70 of different customers
* The 4th attribute is Annual Income in k$ which have different customer have a different Income some have very low some have middle and some have very high income
* The last attribute Spending Score which he spends on Mall.


### Step to do:

        1. Import the necessary packages from scikit-learn (KMeans, AgglomerativeClustering, and silhouette_score). 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

```

```python
mc_df =pd.read_csv('Mall_Customers.csv')
mc_df.tail(10)
```

       2. Assigning the two features Annual Income (k$)" and "Spending Score (1-100)" to X. 

```python
X=mc_df[['Annual Income (k$)', 'Spending Score (1-100)']]
X
```

        3. Visualizing the the two featuerse ("Annual Income (k$)" And "Spending Score (1-100)) to understand the data structure. 

```python
# Visualize the data:
plt.figure(figsize=(10,6))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'])

plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

        4. implementation of k-means on the X features using K= 5. 

```python
kmeans = KMeans(n_clusters=5, random_state=42)

kmeans.fit(X)

cluster_centers = kmeans.cluster_centers_
cluster_labels = kmeans.labels_

print("Cluster Centers:")
print(cluster_centers)
print("\nCluster Labels:")
print(cluster_labels)
```

        5. Plotting the predicted clusters from k-means. 

```python
plt.figure(figsize=(10,6))
mc_df['Cluster'] = cluster_labels


for i in range(5):
    cluster_data = mc_df[mc_df['Cluster'] == i]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {i}', alpha=0.7)

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='x', color='black', s=100, label='Centroids')

plt.title('K-Means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

        6. computing and print the silhouette score of K-mean clustering method 

```python
silhouette_avg = silhouette_score(X, cluster_labels)

print(f"Silhouette Score: {silhouette_avg}")
```

        7. Implementation of Agglomerative Clustering on the X.

 implementing the Agglomerative Clustering three times using three types of linkages (single, complete and average)

#### a - single linkages:
* Implementation of Agglomerative Clustering with single linkages,  (1 point)
* Plot the predicted clusters. (1 point)
* compute the silhouette score  (1 point)

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=5, linkage='single')

agglomerative_clustering.fit(X)

cluster_labels_single = agglomerative_clustering.labels_

print("Cluster Labels (Single Linkage):")
print(cluster_labels_single)
```

```python
# Plot the predicted clusters

mc_df['Cluster'] = cluster_labels_single

plt.figure(figsize=(10, 6))
for i in range(5):
    cluster_data = mc_df[mc_df['Cluster'] == i]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {i}', alpha=0.7)

plt.title('Agglomerative Clustering with Single Linkage')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
# compute and print the silhouette score
silhouette_avg_single = silhouette_score(X, cluster_labels_single)

print(f"Silhouette Score (Single Linkage): {silhouette_avg_single}")
```

#### b - complete linkages:
* Implementation of Agglomerative Clustering with complete linkages,  (1 point)
* Plot the predicted clusters. (1 point)
* compute the silhouette score  (1 point)

```python
# Implementation of Agglomerative Clustering with complete linkages
agglomerative_clustering_complete = AgglomerativeClustering(n_clusters=5, linkage='complete')


agglomerative_clustering_complete.fit(X)

cluster_labels_complete = agglomerative_clustering_complete.labels_


print("Cluster Labels (Complete Linkage):")
print(cluster_labels_complete)
```

```python
# Plot the predicted clusters
mc_df['Cluster'] = cluster_labels_complete

# Create a scatter plot for each cluster
plt.figure(figsize=(10, 6))
for i in range(5):
    cluster_data = mc_df[mc_df['Cluster'] == i]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {i}', alpha=0.7)

plt.title('Agglomerative Clustering with Complete Linkage')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()

```

```python
# compute and print the silhouette score
silhouette_avg_complete = silhouette_score(X, cluster_labels_complete)

print(f"Silhouette Score (Complete Linkage): {silhouette_avg_complete}")
```


### Analysis
Silhouette Score(KMeans): 0.553931997444648

Silhouette Score (Single Linkage): 0.2694896445402342

Silhouette Score (Complete Linkage): 0.5529945955148897

From the above it is observed that K-means and Agglomerative Clustering with Complete Linkage perform better in clustering the data compared to Agglomerative Clustering with Single Linkage.
